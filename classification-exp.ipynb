```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tree.base import DecisionTree
from metrics import *
from sklearn.datasets import make_classification
np.random.seed(42)
```

```python
# Code given in the question
X, y = make_classification(
    n_features=2, n_redundant=0, n_informative=2, random_state=1, n_clusters_per_class=2, class_sep=0.5)

# For plotting
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
train_size = 0.7*len(X)
X_train, y_train = X[:int(train_size)], y[:int(train_size)]
X_test, y_test = X[int(train_size):], y[int(train_size):]
X_train_df = pd.DataFrame(X_train)
y_train_s  = pd.Series(y_train)

X_test_df  = pd.DataFrame(X_test)
y_test_s   = pd.Series(y_test)
print("Train data length: ", len(X_train))
print("Test data length: ", len(X_test))
```

```python
tree = DecisionTree(criterion='gini_index',max_depth=10)
tree.fit(X_train_df, y_train_s)
y_pred = tree.predict(X_test_df)
print("Accuracy: ", accuracy(y_test_s, y_pred))
for cls in sorted(y_test_s.unique()):
    print(f"Class {cls} — Precision: {precision(y_test, y_pred, cls)}, Recall: {recall(y_test, y_pred, cls)}")


```
